### Question 5

`Late` loans are defined as loans that were paid `more than 3 days` after their due date. Please find the average number of days that these loans were paid beyond their due date for each `week of the year`. This will require you to break all `appilcationDate's` down into the week of the year they were made and then perform a group by. 

What is the average number of days that `late` loans applied for in the second week of the year were paid late? Please sumbit your answer rounded to two decimal places. 

```python
df.describe()
```

```python
df.groupby(df["appilcationDate"].dt.week).count().head()
```

```python
df["week_of_the_year"]=df["appilcationDate"].dt.week
df.head()
```

```python
df["RPnegDifferences"]=df["dateDifference"].dt.days
df.head()
```

```python
df.head()
```

```python
df= df[['customerID', 'loanId', 'appilcationDate', 'LoanNumber', 'LoanAmount', 'InterestRate', 'TermDays', 'repaymentDueDate', 'repaymentPaidDate','dateDifference','CountDays','week_of_the_year', 'RPnegDifferences'   ]].copy()
df['RPnegDifferencesA'] = df['RPnegDifferences'].where(df['RPnegDifferences'] > 0)
df['RPnegDifferencesB'] = df['RPnegDifferences'].where(df['RPnegDifferences'] < 0)
df['lateloans'] = df['RPnegDifferencesB'].where(df['RPnegDifferencesB'] < -3)
df.tail(11)
```

```python
df.groupby('week_of_the_year').mean()[['RPnegDifferencesB']]
```

```python
round(df.loc[df['week_of_the_year']==2]['week_of_the_year'].mean(),2)
```

```python
round(df.loc[df['week_of_the_year']==2]['lateloans'].mean(),2)
```
